### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
link = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(link,sep=',')
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index(keys = ['PassengerId'],inplace=True)
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
male_c = (titanic['Sex'] == "male").sum()
female_c = (titanic['Sex'] == "female").sum()

plt.pie(x = [male_c,female_c],labels=["male","female"],autopct="%.3f%%")
plt.title("sex proportion")
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
c = {"male":'red',"female":'blue'}
plt.scatter(y=titanic.Fare,x = titanic.Age,c = titanic.Sex.apply(lambda x : c[x]))
# plt.legend()
plt.show() 
```

```python
import seaborn as sns
sns.scatterplot(data=titanic,y='Fare',x='Age',hue='Sex')
```
